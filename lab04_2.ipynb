```python
# Задание task_04_02_01.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def sum_of_digits(n):
    s = 0
    while n > 0:
        digit = n % 10
        if digit < 5:
            s += digit
        n //= 10
    return s

try:
    n = int(input())
    print(sum_of_digits(n))
except:
    print(0)
```

```python
# Задание task_04_02_02.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def primes(a, b):
    res = []
    for i in range(a, b + 1):
        if i < 2:
            continue
        c = 0
        for j in range(1, i + 1):
            if i % j == 0:
                c += 1
        if c == 2:
            res.append(i)
    return res

try:
    a = int(input())
    b = int(input())
    print(primes(a, b))
except:
    print("Ошибка ввода")
```

```python
# Задание task_04_02_03.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def first_vacant_row(seats):
    max_count = 0
    max_row = 0

    for row_index, row in enumerate(seats):
        available_seats_count = row.count(0)
        if available_seats_count > max_count:
            max_row = row_index
            max_count = available_seats_count

    return (max_row + 1, max_count) if max_count > 0 else (0, 0)

try:
    n = int(input())
    seats = []
    for i in range(n):
        row = list(map(int, input().split()))
        seats.append(row)
    result = first_vacant_row(seats)
    print(f"({result[0]}, {result[1]})")
except:
    print("(0, 0)")
```

```python
# Задание task_04_02_04.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def min_pair(nums):
    min_val = nums[0] + nums[1]
    for i in range(1, len(nums) - 1):
        s = nums[i] + nums[i + 1]
        if s < min_val:
            min_val = s
    return min_val

try:
    nums = [int(x) for x in input().split()]
    if len(nums) < 2:
        print("Ошибка: для поиска суммы нужно как минимум два числа.")
    else:
        print(min_pair(nums))
except ValueError:
    print("Ошибка: введите числа, разделенные пробелом.")
except Exception:
    print("Ошибка ввода")
```

```python
# Задание task_04_02_05.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def non_negatives(nums):

    i = 0
    while i < len(nums):
        if nums[i] < 0:
            del nums[i]
        else:
            i += 1
    return nums

try:
    input_str = input().strip()
    nums = [int(x) for x in input_str.split()]
    result = non_negatives(nums)
    print(result)
except ValueError:
    print("Ошибка: введите числа, разделенные пробелом.")
```

```python
# Задание task_04_02_06.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def unemployment_rate(unemployed, employed):
    return unemployed / (unemployed + employed)

try:
    unemployed = int(input())
    employed = int(input())

    if unemployed < 0 or employed < 0:
        print("Ошибка: Количество людей не может быть отрицательным.")
    elif unemployed + employed == 0:
        print("Ошибка: Общее количество людей (занятые + безработные) не может быть равно нулю.")
    else:
        rate = unemployment_rate(unemployed, employed)
        print("Уровень безработицы = {:.1%}".format(rate))

except ValueError:
    print("Ошибка: Введено нечисловое значение. Пожалуйста, введите целое число.")
```

```python
# Задание task_04_02_07.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def power(x, y=2):
    """Вернуть x^y."""
    if y == 0:
        return 1
    else:
        return x * power(x, y - 1)

try:
    x = int(input())
    y = int(input())

    if y < 0:
        print("Ошибка: функция не предназначена для отрицательных степеней.")
    elif y > 100:
        print("Ошибка: слишком большая глубина рекурсии. Значение 'y' слишком велико.")
    else:
        result = power(x, y)
        print(result)

except ValueError:
    print("Ошибка: введено нечисловое значение. Пожалуйста, введите целые числа.")
except RecursionError:
    print("Ошибка: слишком большая глубина рекурсии. Значение 'y' слишком велико.")
except Exception as e:
    print(f"Произошла непредвиденная ошибка: {e}")
```

```python
# Задание task_04_02_08.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

try:
    n = int(input())
    middle_names = {}
    count = 0

    for i in range(n):
        fio = input().split()
        if len(fio) >= 3:
            middle_name = fio[2]
            middle_names[middle_name] = middle_names.get(middle_name, 0) + 1
            count += 1

    if middle_names:
        print("Самое частое отчество:", sorted(middle_names.items(), key=lambda x: x[1])[-1][0])
    else:
        print("В списке не найдено ни одного человека с отчеством.")

    print("В расчете участвовало человек:", count)

except:
    print("Ошибка: количество человек должно быть целым числом.")
```

```python
# Задание task_04_02_09.
#
# Выполнил: Губайдуллина А.И.
# Группа: АБП-231

def f(x):

    return x**2 / (x + 2) - 3

try:
    k = int(input())
    h = float(input())

    if h <= 0:
        print("Ошибка: шаг табуляции должен быть положительным числом.")
    else:
        print("{:>10} {:>10}".format("x", "f(x)"))
        x = -k
        while x <= k:
            try:
                result = f(x)
                print("{:10.2f} {:10.2f}".format(x, result))
            except:
                print("{:10.2f} {:>10}".format(x, "-"))
            x += h

except ValueError:
    print("Ошибка: введено нечисловое значение. Введите корректные числа.")
```
